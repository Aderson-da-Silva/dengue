```python
import pandas as pd
```

```python
dengue_2022 = pd.read_csv('./data/dengue/DENGBR22.csv')
dengue_2023 = pd.read_csv('./data/dengue/DENGBR22.csv')
dengue_2024 = pd.read_csv('./data/dengue/DENGBR22.csv')
dengue_2025 = pd.read_csv('./data/dengue/DENGBR22.csv')
```

```python
dengue_2023.head(10)
```

```python
dengue_geral = pd.concat([dengue_2022, dengue_2023, dengue_2024, dengue_2025], ignore_index=True)
```

```python
del dengue_2022
del dengue_2023
del dengue_2024
del dengue_2025
```

```python
dengue_geral.info()
```

```python
resumo_colunas = pd.DataFrame({
    'Tipo de Dado': dengue_geral.dtypes,
    'Valores Nulos': dengue_geral.isnull().sum()
})

```

```python
size_df = len(dengue_geral)
```

```python
# Filtra apenas as colunas que têm algum valor nulo para focar no problema
print("Resumo das Colunas (Apenas as que possuem Nulos):")
limite_50_percent = len(dengue_geral) / 2

# 2. Filtra o DataFrame 'resumo_colunas' para mostrar apenas
#    as colunas onde 'Valores Nulos' > limite_50_percent
colunas_com_muitos_nulos = resumo_colunas[resumo_colunas['Valores Nulos'] > limite_50_percent]

# 3. Ordena o resultado pela coluna 'Valores Nulos' do maior para o menor
colunas_muito_esparsas_ordenadas = colunas_com_muitos_nulos.sort_values(
    by='Valores Nulos',
    ascending=False
)

print(colunas_muito_esparsas_ordenadas)
```

```python
for col, dados_linha in colunas_muito_esparsas_ordenadas.iterrows():
    if dados_linha['Valores Nulos'] == size_df:
        dengue_geral.drop(col, axis=1, inplace=True)
        
```
